### <font color=#cebaa4> **Carregando os dados**

````python
# Importa as bibliotecas
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
````

````python
# Carrega os dados
lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}
````

````python
# Transforma os dados em um DataFrame, tornando 'lojas' o index
vendas_2022 = pd.DataFrame(vendas_2022, index=lojas)
````

````python
# Armazena os meses em uma lista
meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
````

### <font color=#cebaa4> **Criando os subplots**

````python
# Cria a figura, configura os espa√ßamentos e o t√≠tulo superior
fig, axs = plt.subplots(2, 2, figsize= (10, 6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Vendas em 2022 por loja')

# Cria e configura o subplot da 'Loja A'
axs[0,0].plot(meses, vendas_2022.loc['A'], color='#f91e52')
axs[0,0].set_title('Loja A')

# Cria e configura o subplot da 'Loja B'
axs[0,1].plot(meses, vendas_2022.loc['B'], color='#4ef1fa')
axs[0,1].set_title('Loja B')

# Cria e configura o subplot da 'Loja C'
axs[1,0].plot(meses, vendas_2022.loc['C'], color='#d89b07')
axs[1,0].set_title('Loja C')

# Cria e configura o subplot da 'Loja D'
axs[1,1].plot(meses, vendas_2022.loc['D'], color='#7a67de')
axs[1,1].set_title('Loja D')

# Configura os t√≠tulos dos eixos dos subplots
for ax in axs.flat:
    ax.set_xlabel('M√™s')
    ax.set_ylabel('Vendas')
    ax.tick_params(axis='x', labelsize=8)
    ax.tick_params(axis='y', labelsize=8)

# Estabelece os limites para o Eixo Y
ymin = 50
ymax = 450
for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

# Exibe o gr√°fico
plt.show()
````
